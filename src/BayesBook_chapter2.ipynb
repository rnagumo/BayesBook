## 0. Uniform Distribution (一様分布)

本には載っていませんが。。。   
stats.uniformで一様分布が指定され、pdf(Probability Density Function, 確率密度関数)で分布が表示される。  
区間[loc, loc+scale]で確率1の分布を作成する。

```python
x = np.arange(-1.0, 3.0, 0.01)
p = stats.uniform.pdf(x, loc=0, scale=2)

plt.plot(x, p)
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('Uniform')
plt.grid()
plt.show()
```

データのサンプリング  
rvs(Random Variates, 確率変数)で、分布に従う確率変数を取得する。

```python
y = stats.uniform.rvs(loc=0, scale=2, size=1000)

plt.plot(x, p)
plt.hist(y, normed=True)
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('Uniform: mean={:.2f}, var={:.2f}'.format(y.mean(), y.var()))
plt.grid()
plt.show()
```

# 離散分布

## 1. ベルヌーイ分布 (Bernoulli Distribution)

$$
\begin{equation*}
Bern(\mu) = \mu^{x}(1 - \mu)^{1-x}
\end{equation*}
$$

$x \in \{a, b\}$を出力する離散分布。  
離散分布なので、pdfの代わりにpmf(Probability mass function, 確率質量関数)を用いる。

```python
x = np.array([0, 1])
p = stats.bernoulli.pmf(x, 0.2)

plt.stem(x, p)
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('Bernoulli')
plt.grid()
plt.show()
```

データサンプリング  
サンプリング結果を、正規化されたヒストグラムで表現している。

```python
y = stats.bernoulli.rvs(p=0.2, size=1000)

plt.stem(x, p)
plt.hist(y, bins=np.linspace(-0.5, 1.5, 3), rwidth=0.5, normed=True, color='g', alpha=0.5)
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('Bernoulli')
plt.grid()
plt.show()
```

## 2. 二項分布 (Binomial Distribution)

$$
\begin{equation*}
Bin(m | M, \mu) = {}_M C_m \mu^{m}(1 - \mu)^{M-m}
\end{equation*}
$$

x <- {0, 1}を出力する試行をM回繰り返したとき、x=1がm回出る回数の確率p。

```python
m = np.arange(0, 20)
p = stats.binom.pmf(k=m, n=20, p=0.2)

plt.stem(m, p)
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('Binomial')
plt.grid()
plt.show()
```

データサンプリング  
サンプリング結果を、正規化されたヒストグラムで表現している。

```python
y = stats.binom.rvs(n=20, p=0.2, size=1000)

plt.stem(m, p)
plt.hist(y, normed=True, color='g', alpha=0.5)
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('Binomial')
plt.grid()
plt.show()
```

## 3. カテゴリ分布 (Categorical Distribution)

$$
\begin{equation*}
Cat(s | \pi) = \prod_{k=1}^{K} \pi_{k}^{s_k}
\end{equation*}
$$


K次元ベクトルsの成分の内、一つだけが1, 残りは0となる分布。  
後述の多項分布において試行回数M=1の場合に対応するので、グラフは割愛。

## 4. 多項分布 (Multinomial Distribution)

$$
\begin{equation*}
Mult(m | \pi, M) = M! \prod_{k=1}^{K} \frac{\pi_{k}^{m_k}}{m_k !}
\end{equation*}
$$

カテゴリ分布の試行をM回繰り返したときにk番目の事象が出現する回数m_kの分布。

以下の例では、3次元ベクトルの多項分布を示している。  
ただし、各次元の出現回数m=x, 総試行回数M=nである。

1. 各次元の出現確率がpiで与えられる。  
2. 総試行回数nを決める。
3. 第1, 第2次元の出現回数x1, x2を[0, n]の範囲で決める。
4. x1 + x2 + x3 = nより、第3次元の出現回数が決まる。
5. このパラメータでの多項分布の確率質量関数を計算する。
6. 第1, 第2次元の出現確率をプロットする。（第3次元は、4.の制約により決定される。）

参考：[Matlab mnpdf](https://jp.mathworks.com/help/stats/mnpdf.html)

```python
# 1. Probability of emerging for each dimension
pi = np.array([1/2, 1/3, 1/6])

# 2. Number of trial
n = 10

# 3. Set the number of the trial falling into each category
X1, X2 = np.meshgrid(np.arange(n), np.arange(n))
X1 = X1.reshape(-1)
X2 = X2.reshape(-1)

# 4. Trial number of the 3rd dimension is calculated
X3 = n - (X1 + X2)

# 5. Get pmf
p = stats.multinomial.pmf(x=np.array([X1, X2, X3]).T, n=n, p=pi)

# 6. Plot for 1st & 2nd dimension
fig = plt.figure()
ax = Axes3D(fig)
bar = ax.bar3d(X1, X2, np.zeros(len(p)), 1, 1, p)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('Probability mass')
plt.title('Multinomial Distribution')
plt.grid()
plt.show()
```

## 5. ポアソン分布 (Poisson Distribution)

$$
\begin{equation*}
Poi(x | \lambda) = \frac{\lambda^{x}}{x!} e^{-\lambda}
\end{equation*}
$$

非負の整数xを生成する離散分布。

```python
x = np.arange(1, 20)
l = 4.0
p = stats.poisson.pmf(x, l)

plt.stem(x, p)
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('Poisson (l={})'.format(l))
plt.grid()
plt.show()
```

データサンプリング  
サンプリング結果を、正規化されたヒストグラムで表現している。

```python
y = stats.poisson.rvs(l, size=1000)

plt.stem(x, p)
plt.hist(y, bins=y.max(), normed=True, color='g', alpha=0.5)
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('Poisson')
plt.grid()
plt.show()
```

# 連続分布

## 5. ベータ分布 (Beta Distribution)

$$
\begin{equation*}
Beta(\mu | a, b) = C_B(a, b) \mu^{a-1} (1-\mu)^{b-1}
\end{equation*}
$$

$\mu \in (0, 1)$となるような変数を生成する確率分布

```python
x = np.arange(0, 1, 0.01)
a = 10.0
b = 2.0
p = stats.beta.pdf(x, a=a, b=b)

plt.plot(x, p)
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('Beta (a={}, b={})'.format(a, b))
plt.grid()
plt.show()
```

## 6. ディリクレ分布 (Dirichlet Distribution)

$$
\begin{equation*}
Dir(\pi | \alpha) = C_D(\alpha) \prod_{k=1}^{K} \pi_k^{\alpha_k-1}
\end{equation*}
$$

ベータ分布を多次元に拡張。  
$\sum_{k=1}^{K}\pi_k=1$かつ$\pi_k \in (0, 1)$となるような変数を生成する確率分布。  
多次元分布を描画するのに断念したので、グラフは割愛。

## 7. ガンマ分布 (Gamma Distribution)

$$
\begin{equation*}
Gam(\lambda | a, b) = C_G(a, b) \lambda^{a-1} e^{-b\lambda}
\end{equation*}
$$

正の実数$\lambda \in R^D$を生成する分布。  
statsではb=1で固定してある。

```python
x = np.arange(0, 20, 0.1)
a = 3.0
p = stats.gamma.pdf(x, a=a)

plt.plot(x, p)
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('Gamma (a={}, b=1.0)'.format(a))
plt.grid()
plt.show()
```

データサンプリング  
サンプリング結果を、正規化されたヒストグラムで表現している。

```python
y = stats.gamma.rvs(a, size=1000)

plt.plot(x, p)
plt.hist(y, normed=True, color='g', alpha=0.5)
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('Gamma (a={}, b=1.0)'.format(a))
plt.grid()
plt.show()
```

## 8. ガウス分布 (Gaussian Distribution)

$$
\begin{equation*}
{\it N}(x | \mu, \sigma^2) = \frac{1}{\sqrt{2\pi\sigma^2}} exp \left(-\frac{(x-\mu)^2}{2\sigma^2}\right)
\end{equation*}
$$

またの名を正規分布という。

```python
x = np.arange(-5, 5, 0.01)
mu = -0.5
std = 1.5
p = stats.norm.pdf(x, loc=mu, scale=std)

plt.plot(x, p)
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('Gaussian (mu={}, var={})'.format(mu, std))
plt.grid()
plt.show()
```

データサンプリング  
サンプリング結果を、正規化されたヒストグラムで表現している。

```python
y = stats.norm.rvs(loc=mu, scale=std, size=1000)

plt.plot(x, p)
plt.hist(y, normed=True, color='g', alpha=0.5)
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('Gaussian (mu={}, var={})'.format(mu, std))
plt.grid()
plt.show()
```

## 10. ウィシャート分布 (Wishart Distribution)

$$
\begin{equation*}
{\it W}(\Lambda | \nu, W) = C_W(\nu, W)|\Lambda|^{\frac{\nu-D-1}{2}}
exp \left(-\frac{1}{2}Tr(W^{-1}\Lambda) \right)
\end{equation*}
$$

$D*D$の正定値行列$\Lambda$を生成する確率分布。  
$\Lambda$は多次元ガウス分布の共分散行列の逆行列である精度行列である。  
下図では、生成された精度行列から生成された正規分布（平均=[0, 0]）のある値の大きさでの等高線を引いている。

```python
x, y = np.meshgrid(np.arange(-5, 5, 0.01), np.arange(-5, 5, 0.01))
pos = np.dstack((x, y))

w = np.matrix([[2.0, 0], [0, 2.0]])
covs = stats.wishart.rvs(df=5, scale=w, size=10)

for cov in covs:
    z = stats.multivariate_normal.pdf(pos, mean=[0, 0], cov=cov)
    plt.contour(x, y, z, levels=[0.01])
    
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('Gaussian')
plt.grid()
plt.show()
```
